```python
import pandas as pd
```

```python
df=pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/TheTitanic.csv')
```

```python
df.head()
```

```python
df.columns=df.iloc[0]
df=df[1:]
```

```python
df
```

```python
df.info()
```

```python
df.columns=['sn','pclass','survived','Name','gender','age','family','fare','embarked','date']
```

```python
df.columns
```

```python
df.isnull().sum()
```

**# Fill missing values for 'Age' with the median**
df['Age'].fillna(df['Age'].median(), inplace=True)

**# Drop rows with missing values in 'Embarked' (if any)**
df.dropna(subset=['Embarked'], inplace=True)

**# Drop 'Cabin' column due to high number of missing values**
df.drop(columns=['Cabin'], inplace=True)

```python
# drop unnecessary columns
df.drop(['Name','sn','date'], axis=1, inplace=True)
```

```python
# fill missing values
import numpy as np
df['age'].fillna(df['age'].mean(),inplace=True)
df['embarked'].fillna(df['embarked'].mode(), inplace=True)
```

```python
df['embarked'].value_counts()
```

```python
# split data into train and test sets
from sklearn.model_selection import train_test_split
X = df.drop('survived', axis=1)
y = df['survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# print the shapes of the train and test sets
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```
